```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
prices = pd.read_csv("stock.csv", parse_dates = ["Date"])
# prices["Date"] = pd.to_datetime(prices["Date"])
```

```python
prices.head()
```

```python
stock = pd.DataFrame()
```

```python
stock["Date"] = prices["Date"]
stock["Apple"] = prices["Adj Close Price"]
```

```python
stock.head(40)
```

```python
stock["Apple"].iloc[:30].mean(), stock["Apple"].iloc[1:31].mean()
```

```python
# short term simple moving average
stock["SM30"] = stock["Apple"].rolling(30).mean()

# long term simple moving average
stock["SM100"] = stock["Apple"].rolling(100).mean()
```

```python
stock.head()
```

```python
plt.figure(figsize=(16,5))

plt.plot(stock["Date"],stock["Apple"], alpha = 0.7, label = "Apple Stock Price")
plt.plot(stock["Date"],stock["SM30"], alpha = 0.6, label = "SM30")
plt.plot(stock["Date"],stock["SM100"], alpha = 0.6, label = "SM100")

plt.title("Apple Historical Adj Close Price")
plt.xlabel("From 27 May, 2014 to 22 May, 2020")
plt.ylabel("Adj Close Price")
plt.legend()

plt.show()
```

```python
def buy_sell(df):
    buy_signal_price = []
    sell_signal_price = []
    holding = [] # 1 for hold, 0 for not hold
    actions = [] # 0 : wait, 1: Buy, -1: Sell
    sig = 0
    for i,row in stock.iterrows():
        if row["SM30"] > row["SM100"]: # 91.173741 > 87.780364
            if sig != 1: # 1 != 1
                buy_signal_price.append(row["Apple"])
                sell_signal_price.append(np.nan)
                holding.append(1)
                sig = 1
                actions.append(sig)
            else:
                buy_signal_price.append(np.nan)
                sell_signal_price.append(np.nan)
                holding.append(1)
                actions.append(0)
        
        elif row["SM30"] < row["SM100"]:
            if sig != -1: # -1 != -1
                buy_signal_price.append(np.nan)
                sell_signal_price.append(row["Apple"])
                holding.append(0)
                sig = -1
                actions.append(sig)
            else:
                buy_signal_price.append(np.nan)
                sell_signal_price.append(np.nan)
                holding.append(0)
                actions.append(0)
        else:
            buy_signal_price.append(np.nan)
            sell_signal_price.append(np.nan)
            holding.append(0)
            actions.append(0)
    return buy_signal_price, sell_signal_price, holding, actions    
```

```python
buysell =  buy_sell(stock)
```
